## Programmatic access — top-5 features per sector at h=6

```python
from src.config import FF_INDUSTRIES, SHAP_DIR, FEATURES_DIR
X = pd.read_parquet(FEATURES_DIR / 'X_v1.parquet')

rows = []
for s in FF_INDUSTRIES:
    sv = np.load(SHAP_DIR / f'shap_values_{s}_h6.npy')
    mean_abs = np.abs(sv).mean(axis=0)
    top5 = np.argsort(mean_abs)[-5:][::-1]
    rows.append({'sector': s,
                 **{f'rank_{i+1}': X.columns[idx] for i, idx in enumerate(top5)}})
pd.DataFrame(rows)
```
